```python
import pandas as pd
```

```python
# define a 1D series
mySeries=pd.Series([1,2,3,4,5],index=['row1','row2','row3','row4','row5'])
mySeries
```

```python
# To print values in the defined series 
mySeries.values
```

```python
# To print the index in the defined series 
mySeries.index
```

```python
# To show the value in the specific index 
mySeries.row2
```

```python
# Another way to show the value in the specific index 
mySeries['row2']
```

```python
# To show some part of a series
mySeries[mySeries>3]
```

```python
# To rename a series by a dictionary
mySeries.rename({'row1':'a', 'row2':'b', 'row3':'c', 'row4':'d', 'row5':'e'})
```

```python
# define a dataframe --Approach 1
myArrey=np.array([[1,5,9,13],[2,6,10,14],[3,7,11,15],[4,8,12,16]])
myDf=pd.DataFrame(myArrey,index=['row1', 'row2', 'row3', 'row4']
                  ,columns=['col1','col2','col3','col4'])
myDf
```

```python
# Define a dataframe by a dictionary ---Approach 2
myDict={'col1':[1,2,3,4],'col2':[5,6,7,8],'col3':[9,10,11,12],'col4':[13,14,15,16]}
myDf2=pd.DataFrame(myDict,index=['row1', 'row2', 'row3', 'row4']
                  ,columns=['col1','col2','col3','col4'])
myDf2
```

```python
# To show dataframe index
myDf.index
```

```python
# To show dataframe columns
myDf.columns
```

```python
# To show dataframe value
myDf.values
```

```python
# To print a specific value in a dataframe
myDf.loc['row1']['col2']
```

```python
# # To print a specific value in a dataframe by its location
myDf.iloc[0][1]
```

```python
# Change one column in a dataframe
myDf['col5']=[20,21,22,23]
myDf
```

```python
# Replace some values in a dataframe
myDf.loc[['row1','row2'],'col1']=0
myDf
```

```python
# Reset row names for a dataframe
myDf.reset_index(drop=True,inplace=True)
myDf
```

```python
# Remove col5 form a dataframe
myDf.drop('col5',axis=1,inplace=True)
myDf
```

```python
# Rename col4 with c4
myDf.rename(columns={'col4':'c4'},inplace=True)
myDf
```

```python
# Replace all zeros in dataframe with one
myDf=myDf.replace(0,1)
myDf
```

```python
# Write a dataframe with float values----Approach 1
myDf.col1=['{:.2f}'.format(x) for x in myDf.iloc[:,0]]
myDf
```

```python
# Write a dataframe with float values----Approach 2
myDf['col2']=myDf['col2'].apply(lambda x:'{:.2f}'.format(x))
myDf
```

```python
# Sort dataframe in terms of columns---- to sort in terms of columns try axis=0 
myDf.sort_index(axis=1,ascending=False)
```

```python
# Sort dataframe in terms of rows---- to sort in terms of columns try axis=1 
myDf.sort_values(by='col1',ascending=False)
```

```python
# To show the head of a dataframe --- the default is 5
myDf.head(2)
```

```python
# To show the tail of a dataframe --- the default is 5
myDf.tail(2)
```
