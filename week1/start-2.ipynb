```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as dt
```

```python
iris=dt.load_iris()
#print(iris)
x=iris.data
y=iris.target
#print(x.shape)

FN=iris.feature_names
#print(FN)
TN=iris.target_names
print(TN)

print(x[:10])
print(y[:10])
```

```python
# plot iris dataset
plt.figure()
plt.scatter(x[:,1],x[:,2],s=15,c=y)
#plt.scatter(x[:,2],x[:,3],s=15,c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('iris dataset')
#plt.legend(['setosa', 'versicolor', 'virginica'], loc ="upper right")
plt.show()
```

```python
# Dividing our dataset to trian and test set
import sklearn.model_selection as ms
#Tr_x, Tes_x, Tr_y, Tes_y=ms.train_test_split(x,y,train_size=0.8)
Tr_x, Tes_x, Tr_y, Tes_y=ms.train_test_split(x,y,train_size=0.8, random_state=42)
print(Tr_x.shape)
print(Tes_x[20:])
print(Tes_y[20:])
```

```python
# Scale our dataset (preprocessing)
import sklearn.preprocessing as pp
MMS = pp.MinMaxScaler(feature_range=(0, +1))
MMS.fit(Tr_x)
Tr_x1=MMS.transform(Tr_x)
print(Tr_x1)
Tes_x1=MMS.transform(Tes_x)
print(Tes_x1)
```

```python
plt.figure()
plt.scatter(Tr_x1[:,0],Tr_x1[:,1],s=15,c=Tr_y)
plt.xlabel('Train_X1')
plt.ylabel('Train_X2')
plt.title('Scaled train iris dataset')
plt.show()
```
